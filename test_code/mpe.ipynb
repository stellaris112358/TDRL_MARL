```python
from pettingzoo.mpe import simple_spread_v3

env = simple_spread_v3.parallel_env(render_mode="human",dynamic_rescaling = True)
observations, infos = env.reset()

while env.agents:
    # this is where you would insert your policy
    actions = {agent: env.action_space(agent).sample() for agent in env.agents}

    observations, rewards, terminations, truncations, infos = env.step(actions)
env.close()
```

```python
class CustomSpreadEnv:
    def __init__(self, **kwargs):
        self.env = simple_spread_v3.parallel_env(**kwargs)
        self.agents = self.env.agents
        self.action_space = self.env.action_space
        self.observation_space = self.env.observation_space
    
    def reset(self, seed=None, options=None):
        observations, infos = self.env.reset(seed=seed, options=options)
        self.agents = self.env.agents
        return observations, infos
    
    def step(self, actions):
        # 在这里添加你的自定义逻辑
        observations, rewards, terminations, truncations, infos = self.env.step(actions)
        
        # 示例：修改奖励
        # rewards = {agent: reward * 2 for agent, reward in rewards.items()}
        
        self.agents = self.env.agents
        return observations, rewards, terminations, truncations, infos
    
    def close(self):
        self.env.close()

# 使用自定义环境
custom_env = CustomSpreadEnv(render_mode="human", dynamic_rescaling=True)
observations, infos = custom_env.reset()

while custom_env.agents:
    actions = {agent: custom_env.action_space(agent).sample() for agent in custom_env.agents}
    observations, rewards, terminations, truncations, infos = custom_env.step(actions)

custom_env.close()
```
